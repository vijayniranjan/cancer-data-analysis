```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
```

```python
datax = pd.read_csv('C:/Users/vijay niranjan/Desktop/braincancer/brainx.csv')
datay = pd.read_csv('C:/Users/vijay niranjan/Desktop/braincancer/brainy.csv')
combine = [datax, datay]
```

```python
datax.info()
datay.info()
```

```python
x = datax.iloc[:,0:5598]
y = datay.iloc[:,-1] 
```

```python
datax.head()
```

```python
from numpy import set_printoptions
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=f_classif, k=10)
fit = test.fit(x, y)

set_printoptions(precision=2)
print(fit.scores_)
features = fit.transform(x)
features = pd.DataFrame(fit.scores_)
featurescolumns = pd.DataFrame(x.columns)
featureScores = pd.concat([featurescolumns,features],axis=1)
featureScores.columns = ['features','Score']  
print(featureScores.nlargest(10,'Score'))
```
